```python
import json
import glob
import re
import pandas as pd
import os
import git
```

```python
def get_git_data(project_name):
    project_path = "/home/jovyan/work/projects/"+project_name+"_Bug_1"
    if project_name == "Collections":
        project_path = "/home/jovyan/work/projects/"+project_name+"_Bug_25"
    repo = git.Repo(project_path)
    
    last_commit = repo.rev_parse('origin/HEAD')
    commits = list(repo.iter_commits(last_commit))
    first_commit = commits[-1]
    n_commits = len(commits)
    
    return last_commit, first_commit, n_commits
```

```python
projects = [
    "JacksonXml", "Time", "Collections", "Compress", "Csv", "JacksonCore", "JacksonDatabind", "Gson", "Jsoup",
    "Lang", "Math", "Closure", "Mockito", "Cli", "JxPath", "Codec"
]
projects.sort()
project_list = []
for project in projects:
    last_commit, first_commit, n_commits = get_git_data(project)
    project_list.append({ 
        "name": project,
        "n_bugs": len(os.listdir("/home/jovyan/work/configFiles/"+project+"/bugs")),
        "n_commits": n_commits,
        "first_commit": first_commit.committed_datetime.strftime('%Y-%m-%d'),
        "last_commit": last_commit.committed_datetime.strftime('%Y-%m-%d')
    })
pd.DataFrame(project_list)
```
