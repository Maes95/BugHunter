```python
import json
import glob
import re
import pandas as pd
import os
```

```python
root="/home/jovyan/work"
analysis_results_path = root + "/analysis/results/"
```

```python
real_bics = pd.read_csv('groundOfTruth.csv').to_dict('records')
```

```python
results = []
for bic in real_bics:
    bic_hash = bic['Inducing Change']
    result_path = "{path}/{project}/Bug_{id}/bug_result.json".format(path=analysis_results_path, project=bic['Project'], id=bic['Bug ID'])
    if os.path.isfile(result_path):
        with open(result_path) as f:
            result = json.load(f)
            if result['category'] == "A regression is detected":
                match = False
                for candidate in result['BIC_candidates']:
                    candidate_hash = candidate[1]
                    if candidate_hash == bic_hash: 
                        match = True
                        break
                results.append({
                    'id': result['id'],
                    'real_bic': bic_hash,
                    'match': match
                })
results_df = pd.DataFrame(results)
```
